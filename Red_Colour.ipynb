```python
#import the libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#import image
img = cv2.imread('C:/Users/Asus123/Desktop/Projects/1.jpg')
```

```python
#praper the image and show
plt.figure(figsize=(20,8))
plt.imshow(img)
```

```python
#Convert the BGR format image to RGB format
grid_RGB =cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
#praper the image(RGB) and show
plt.figure(figsize=(20,8))
plt.imshow(grid_RGB)
```

```python
#Convert the RGB format image to HSV format
grid_HSV =cv2.cvtColor(grid_RGB, cv2.COLOR_RGB2HSV)
```

```python
#Specify the lower and upper HSV color (array) value to detect
#for RED
lower = np.array([0,150,50])
upper = np.array([10,255,255])
```

```python
mask = cv2.inRange(grid_HSV, lower,upper)
```

```python
#praper the image(RGB) and show only the red color part
plt.figure(figsize=(20,8))
plt.imshow(mask)
```

```python
res = cv2.bitwise_and(grid_RGB, grid_RGB, mask=mask)
```

```python
#praper the image(RGB) and show only the red part in that image
plt.figure(figsize=(20,8))
plt.imshow(res)
```
